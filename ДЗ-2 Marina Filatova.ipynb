Корпус новостей с сайта новостного издания https://theblueprint.ru/. 


Что нам предстоит сделать:
- достать все заголовки новостей в главной страницы + текст каждой новости
- сохранить в датафрейм с колонками "источник", "дата", "заголовок", "текст"

- сохранить датафрейм в файл
- положить код и получившийся файл в свой репозиторий на гитхабе

```python
import requests as rq
from bs4 import BeautifulSoup
```

```python
url = 'https://theblueprint.ru'
```

```python
page = rq.get(url)

page
```

```python
soup = BeautifulSoup(page.text, 'html.parser')
```

#### Заголовки новостей на сайте есть с тегами a.title, a.news_item. Достанем все заголовки главной страницы с тегом a.title: 

```python
soup.find_all('a', class_='title')
```

Достанем заголовки по ссылкам с тегом a.news_item.

```python
soup.find_all('a', class_='news_item')
```

#### Далее будем доставать текст каждой новости. 

```python
# soup.find('a', class_='title').get('href')
   
```

```python
d = {}
```

Найдем ссылку и заголовок новостей.

```python
result = soup.find_all('a', class_='news_item')

for url_news in result:
    d[url + url_news.get('href').replace(' ', '%20')] = url_news.find('p', class_='title').text
print(d)
```

```python
# soup.find_all('a', class_='title')
```

```python
result2 = soup.find_all('a', class_='title')

for url_title in result2:
    d[url + url_title.get("href").replace(' ', '%20')] = url_title.text.strip()
    
print(d)
```

Создадим словарь, в котором будут ссылка, заголовок, дата и текст новостей.

```python
dict = {}

for idx, key in enumerate(d):
#     print(key)
    childpage = rq.get(key)
    soup = BeautifulSoup(childpage.text, 'html.parser')
    newsdate = soup.find('p', class_ = 'page-article-news__date')
    if newsdate == None:
        newsdate = soup.find('p', class_ = 'date')
    editors = soup.find_all('div', class_='editor')
    newstext = ''
    for editor in editors:
        ps = editor.find_all('p')
        for p in ps:
            spans = p.find_all('span')
            for span in spans:
                newstext = newstext + span.text.strip()
#                 print(span.text.strip())
    dict[idx] = [key, d[key], newsdate.text, newstext]
print(dict)
```

Создадим датафрейм из словаря.

```python
import pandas as pd

            
my_dataframe = pd.DataFrame.from_dict(dict,columns=["источник","заголовок","дата","текст"], orient="index")

print(my_dataframe)
```

Сохраним датафрейм в файл.
